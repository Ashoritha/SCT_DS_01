## Renaming Columns in DataFrame with 


```python
df.rename(columns = {'Country Name':'Country_Name', 'Country Code':'Country_Code'}, inplace=True)
df.head()
```

## Visualizing Distributions of Columns 


```python
for col in df:
    if col != 'Country_Name' and col != 'Country_Code':
        sns.histplot(x=col, data=df, kde=True)
        plt.show() 
```

## Dropping, Renaming, and Transposing the DataFrame for Yearly Data


```python
df_T = df.drop('Country Code', axis=1).rename(columns={'Country Name': 'Year'}).T
df_T.columns = df_T.iloc[0]
df_T = df_T.drop(df_T.index[0]).reset_index(drop=True)
df_T.head()
```

## Visualizing the Transposed DataFrame with Histograms and Line Plots


```python

if 'df_T' in locals():
    for col in df_T.columns:
        plt.figure(figsize=(10, 8))
        sns.histplot(df_T[col], kde=True)
        plt.show()

        plt.figure(figsize=(10, 8))
        sns.lineplot(data=df_T[col])
        plt.show()
```
